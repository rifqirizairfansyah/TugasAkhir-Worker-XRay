```python
import os
from os.path import splitext
import pandas as pd
from os import listdir
from os.path import isfile, join

def extract_model_dataset_name(name_pkl) :
    split_name_pkl = splitext(name_pkl)[0]
    remove_word_1 = split_name_pkl.replace("valid_","")
    split_with = remove_word_1.split("_with_")
    model_name = split_with[0] 
    model_name_split = model_name.split('_')
    del model_name_split[0]
    model_name_string = '_'.join(model_name_split)
    valid_name = split_with[1]
    return model_name_string, valid_name

def open_pkl_file(path):
    with open(path, 'rb') as f:
        data = pickle.load(f)
    avg_auc = data[0]
    pathologies_auc =data[1]
    return avg_auc, pathologies_auc

def process_data(avg_auc, pathologies_auc) :
    data_append = np.append(pathologies_auc, avg_auc)
    return data_append

```

```python
data_auc_model = pd.DataFrame({"Label AUC": pathologies_label})
```

```python

pkl_files = [f for f in listdir(base_pkl_path) if isfile(join(base_pkl_path, f))]
pkl_files=sorted(pkl_files)

for pkl_file in pkl_files :
    print(pkl_file)
    model, valid = extract_model_dataset_name(pkl_file)
    path_pkl = base_pkl_path + pkl_file
    #print(path_pkl)
    avg_auc, pathologies_auc = open_pkl_file(path_pkl)
    #print(avg_auc)
    data_result = process_data(avg_auc, pathologies_auc)
    #print(data_result[18])
    name_label = model + "_with_" + valid
    #print(name_label)
    data_auc_model[name_label] = data_result
    
```
